```python
br_1_7098 = pd.read_pickle('brain_states_1_S7098.pkl')
br_1_7098
```

```python
print("There are " + str(len(br_1_7098)) + " epochs.")
```

```python
# Add the samples to the dataframe
start_time = [t * 250.4 for t in br_1_7098.start_epoch]
end_time = [t * 250.4 for t in br_1_7098.end_epoch]
br_1_7098['start_time'] = start_time
br_1_7098['end_time'] = end_time

# Add the number of epochs to the dataframe
epochs2 = []
for i in range(len(br_1_7098)):
    e = "epoch" + str(i + 1)
    epochs2.append(e)
br_1_7098['epochs'] = epochs2
```

```python
br_1_7098 
```

```python
labeled_data = pd.merge(df, br_1_7098, on ='epochs')
```

```python
labeled_data.drop(['start_epoch', 'end_epoch', 'start_time', 'end_time'], inplace=True, axis=1)
```

```python
labeled_data # The recording with samples labeled with corresponding sleep stages
```

```python
d = []
channels = ['ch1', 'ch2', 'ch3', 'ch4', 'ch5', 'ch6', 'ch7', 'ch8', 'ch9', 'ch10', 'ch11', 'ch12', 'ch13', 'ch14', 'ch15', 'ch16']

for epoch in br_1_7098.epochs:
    e = labeled_data[labeled_data.epochs == epoch]
    #print(epoch)
    a = []
    for channel in channels:
        arr = e[channel].to_list()
        a.append(arr)
    d.append(a)
    print(epoch)
```

```python
df_d = pd.DataFrame(d, columns=channels)    
```

```python
df_d
```

```python
final = pd.concat([br_1_7098, df_d], axis=1)
```
